```python
!pip install PyPDF2
import PyPDF2
import os
```

```python
merger = PyPDF2.PdfMerger()
```

```python
lista_arquivos = os.listdir('/content/drive/MyDrive/Colab Notebooks/Arquivos/Automacao')
lista_arquivos.sort()
print(lista_arquivos)
```

```python
for arquivo in lista_arquivos:
    if ".pdf" in arquivo:
      merger.append(f'/content/drive/MyDrive/Colab Notebooks/Arquivos/Automacao/{arquivo}')
```

```python
pasta_destino = '/content/drive/MyDrive/Colab Notebooks/Arquivos/Automacao/Pdf Final'
if not os.path.exists(pasta_destino):
    os.makedirs(pasta_destino)

arquivo_destino = os.path.join(pasta_destino, 'Pdf Final.pdf')
```

```python
merger.write(arquivo_destino)  # Supondo que "merger" seja o objeto que gera o PDF

print(f"PDF salvo em: {arquivo_destino}")
```
