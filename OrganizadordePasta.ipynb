```python
lista_arquivos = os.listdir(caminho)
print(lista_arquivos)
```

```python
locais = {
    "imagens": [".png", ".jpg", ".jpeg", ".gif", ".bmp", ".tiff"],
    "planilhas": [".xlsx", ".xls", ".csv", ".ods"],
    "pdfs": [".pdf"],
    "documentos": [".docx", ".doc", ".txt", ".rtf", ".odt"],
    "apresentacoes": [".pptx", ".ppt", ".odp"],
    "zips-executaveis": [".zip", ".rar", ".7z", ".tar", ".gz", ".exe"],
    "audios": [".mp3", ".wav", ".aac", ".flac", ".ogg"],
    "videos": [".mp4", ".avi", ".mkv", ".mov", ".wmv"],
    "imagens_vetoriais": [".svg", ".eps", ".ai", ".psd"],
    "fontes": [".ttf", ".otf", ".woff", ".woff2"],
    "scripts": [".py", ".js", ".html", ".css", ".php", ".java", ".c", ".cpp"],
    "bases_de_dados": [".sql", ".db", ".sqlite", ".mdb", ".accdb"],
    "logs": [".log"],
    "outros": []  # Use esta categoria para arquivos que não se encaixam em nenhuma outra
}
```

```python
# Loop através de cada arquivo em 'lista_arquivos'
for arquivo in lista_arquivos:
    # Divide o nome do arquivo em duas partes: o nome e a extensão
    # 'os.path.splitext' retorna uma tupla (nome, extensão)
    nome, extensao = os.path.splitext(f"{caminho}/{arquivo}")

    # Loop através de cada pasta em 'locais'
    for pasta in locais:
        # Verifica se a extensão do arquivo está na lista de extensões associadas à pasta
        if extensao in locais[pasta]:
            # Se a pasta ainda não existe no caminho, ela é criada
            if not os.path.exists(f"{caminho}/{pasta}"):
                os.mkdir(f"{caminho}/{pasta}")

            # Move o arquivo atual para a pasta correspondente
            # 'os.rename' é usado para mover o arquivo para o novo local
            os.rename(f"{caminho}/{arquivo}", f"{caminho}/{pasta}/{arquivo}")
```
