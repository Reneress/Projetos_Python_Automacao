```python

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pandas_datareader import data as pdr
import yfinance as yf
```

```python
ativos = ["ITUB3.SA","VALE3.SA","PETR3.SA","^BVSP","AAPL","MSFT","GOOGL"]


data_i = "2023-01-01"
data_f = "2024-12-31"
resultados = {}

# Obter dados históricos de cada ativo
for ativo in ativos:
    ticker = yf.Ticker(ativo)
    cotacoes = ticker(start=data_i, end=data_f)
    resultados[ativo] = cotacoes["Close"]  # Apenas o preço de fechamento

# Plotar os dados
plt.figure(figsize=(10, 6))  # Tamanho do gráfico
for ativo, precos in resultados.items():
    plt.plot(precos.index, precos, label=ativo)  # Adicionar cada ativo ao gráfico

# Customizar o gráfico
plt.title("Preços de Fechamento dos Ativos", fontsize=16)
plt.xlabel("Data", fontsize=12)
plt.ylabel("Preço de Fechamento (USD)", fontsize=12)
plt.legend(title="Ativos", fontsize=10)
plt.grid(True, linestyle="--", alpha=0.6)

# Exibir o gráfico
plt.show()
```
